```python
import matplotlib.pyplot as plt

# Eje X (24 periodos)
periodo = list(range(1, 25))

# Eje Y (valores aproximados segn tu descripcin)
valor = [
    10, 9, 15, 11, 14, 16, 9, 10, 11, 12, 13, 20,
    22, 26, 25, 31, 27, 27, 27, 24, 26, 30, 27, 34
]

# Crear la figura
plt.figure(figsize=(8, 5))
plt.plot(periodo, valor, marker='o', color='royalblue', linewidth=2)

# Ttulo igual al de la imagen
plt.title('Lneas: serie temporal con tendencia y estacionalidad', fontsize=12, fontweight='bold')

# Etiquetas
plt.xlabel('Periodo')
plt.ylabel('Valor')

# Cuadrcula
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Lmites de los ejes
plt.xlim(0, 25)
plt.ylim(5, 35)

# Mostrar la grfica
plt.show()
```
